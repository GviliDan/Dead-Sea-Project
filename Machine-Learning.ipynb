```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
all_DataFile = pd.read_csv("all_DatasetFile.csv")
all_DataFileByMonth = pd.read_csv("all_DatasetFileByMonth.csv")
```

```python
df_status=all_DataFile.copy()
df_status["Status"] = (df_status["sns_drt"] > 400) & (df_status["tmp_air_max"] > 18)
df_status= df_status.drop(['sns_drt','tmp_air_max', 'Date','DeadSeaLevelsInPrecent','rpr_hail','rpr_snow','rpr_gale'],axis=1)
df_status=df_status.dropna()
df_status
```

```python
df_status.info()
```

```python
TRAINING_FEATURES = df_status.columns[df_status.columns != 'Status'] 
TARGET_FEATURE    = 'Status'
X = df_status[TRAINING_FEATURES]
y = df_status[TARGET_FEATURE]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
clf_model = LogisticRegression().fit(X_train, y_train) #רגרסיה לוגיסטית
y_pred_train=clf_model.predict(X_train)
y_pred=clf_model.predict(X_test)
score = clf_model.score(X_test, y_test) *100
print("train_accuracy = ",accuracy_score(y_true=y_train, y_pred=y_pred_train)) # מציאת הדיוק של אחוז מבדיקת המידע
print("test_accuracy = ",accuracy_score(y_true=y_test, y_pred=y_pred)) # מציאת הדיוק של אחוז בחינת המידע
print(score)  # מציאת הדיוק של החיזוי באחוזים
```
