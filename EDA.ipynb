```python
import seaborn as sns
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import missingno as msno
from numpy import nansum
from numpy import nanmean
```

```python
all_DataFile = pd.read_csv("all_DatasetFile.csv")
all_DataFileByYear = pd.read_csv("all_DatasetFileByYear.csv")
all_DataFileByMonth = pd.read_csv("all_DatasetFileByMonth.csv")
```

```python
yearStart=1976
yearEnd=2021

all_DataFileByYear[(all_DataFileByYear['Year']>=yearStart)&(all_DataFileByYear['Year']<=yearEnd)].plot(
    x='Year',
    y=["DeadSea_Level_End_Of_Year"],
    figsize=(40, 5),
    title='Dead-Sea Levels',
    grid=True
)


all_DataFileByYear[(all_DataFileByYear['Year']>=yearStart)&(all_DataFileByYear['Year']<=yearEnd)].plot(
    x='Year',
    y=['tmp_air_max'],
    title='Average maximum temperature value relative to years',
    figsize=(40, 5),
    grid=True
)

all_DataFileByYear[(all_DataFileByYear['Year']>=yearStart)&(all_DataFileByYear['Year']<=yearEnd)].plot(
    kind='bar',
    x='Year',
    y=['Change_From_Last_Year'],
    title='CHANGE FROM LAST YEAR',
    figsize=(40, 5),
    grid=True
)
```

```python
#Multiplotting
plt.figure(1,figsize=[14,12])

# Figure num 1
plt.subplot(2,2,1)
sns.boxplot(data=all_DataFileByMonth,y="Change_From_Last_Month",x="Month")
plt.title("     ")
plt.xlabel(" ")
plt.ylabel("   ")
plt.ylim([-1.5,1.5])

#figure num 2
plt.subplot(2,2,2)
sns.boxplot(data=all_DataFileByMonth,y="sns_drt",x="Month")
plt.title("      ")
plt.xlabel(" ")
plt.ylabel("   ")
```
